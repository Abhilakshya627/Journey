```python
def linear_regression_model(X,w,b):
    function_x=np.zeros(m)
    for i in range(m):
        function_x[i]=w*X[i]+b
    return function_x
        
        
    
    
```

```python
#calculate cost function

def calculate_cost_function(X,Y,w,b):
    
    sum_cost=0
    for i in range(m):
        function_x=w*X[i]+b
        sum_cost=sum_cost+(function_x-Y[i])**2
    final_cost=(1/(2*m))*sum_cost
    return  final_cost   
```

```python
w=100
b=100
j_cost_cal=calculate_cost_function(X,Y,w,b)
```

```python
print(j_cost_cal)
```

```python
def calculate_gradient(X,Y,w,b):
    dw,db=0,0
    
    for i in range(m):
        function_x=w*X[i]+b
        dw=dw+(function_x-Y[i])*X[i]
        db=db+(function_x-Y[i])
    dw=dw/m
    db=db/m
    
    return dw,db
        

        
    
```

```python
w=100
b=100
j_dw,j_db=calculate_gradient(X,Y,w,b)
print(j_dw,j_db)
```

```python
def gradient_descent_function(X,Y,w1,b1,calculate_gradient,calculate_cost_function,learning_rate,iteration):
    w,b=w1,b1
    for i in range(iteration):
        j_dw,j_db=calculate_gradient(X,Y,w,b)
        w=w-learning_rate*j_dw
        b=b-learning_rate*j_db
        total_jcost=calculate_cost_function(X,Y,w,b)
        print(f"cost at iteration {i}={total_jcost}")
        
    return w,b
    
    
```

```python
w1,b1=0,0
learning_rate=0.001
iteration=1000
w,b=gradient_descent_function(X,Y,w1,b1,calculate_gradient,calculate_cost_function,learning_rate,iteration)
print(f"w:{w} and b:{b}")
```

```python
words=list(map(str.lower,words))
```

# Remove DUPLICATES

```python
words=list(set(words))
```

```python
words
```

# Remove puntuation and numbers

```python
alpha_words=[]

for word in words:
    if word.isalpha():
        alpha_words.append(word)
        
```

```python
alpha_words
```

```python
from nltk.corpus import stopwords
```

```python
filter_words=[]
for word in alpha_words:
    if word not in stopwords.words('english'):
        filter_words.append(word)
        
```
