```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Housing.csv')
df
```

```python
x=list(df['area'])
y=list(df['price'])
m=len(x) 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0) #splitting the dataset into training(80%) and testing(20%) data
```

```python
lr=LinearRegression()
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1)) #fitting the model
```

```python
y_test_pred=lr.predict(np.array(x_test).reshape(-1,1)) #predicting the values of y_test
y_train_pred=lr.predict(np.array(x_train).reshape(-1,1)) #predicting the values of y_train
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,y_train_pred,color='blue')
plt.title('Training Data')
plt.xlabel('Area')
```
